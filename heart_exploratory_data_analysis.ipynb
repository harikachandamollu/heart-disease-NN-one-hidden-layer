### Data Analysis

Explore the dataset to understand the data.

### About data set

This dataset contains 14 attributes, target(14th) variable is the final output variable which depicts the chance(0/1) of heart attack(binary classification).  

### Attribute Information
1. age: age(years)
2. sex: sex(1=male; 0=female)
3. cp: chest pain type(4 values)
4. trestbps: resting blood pressure(int(in mm Hg))
5. chol: serum cholestoral in mg/dl(int)
6. fbs: fasting blood sugar > 120 mg/dl(1=true, 0=false)
7. restecg: resting electrocardiographic results(values 0, 1, 2)
8. thalach: maximum heart rate achieved(int)
9. exang: exercise induced angina(1=yes, 0=no)
10. oldpeak: oldpeak = ST depression induced by exercise relative to rest(float)
11. slope: the slope of the peak exercise ST segment(Values 1,2,3)
12. ca: number of major vessels (0-3) colored by flourosopy
13. thal: 0 = normal; 1 = fixed defect; 2 = reversable defect
14. target: 0= less chance of heart attack 1= more chance of heart attack

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('heart.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['target'].value_counts()
```

**There are 165 patients with heart disease and 138 patients without heart disease.**

```python
%matplotlib inline
#Plot Histogram to find out data distributions of the variables
data.hist(bins=50, figsize=(20,15))
plt.show()
```

**To visualize a distribution, histogram is the most common approach. The count of observations falling within each bin is shown using the height of the corresponding bar.**

```python
#To find correlations between all the possible variables
corr_matrix=data.corr()
corr_matrix
```

```python
#To grab quick correlation, colors provides best results, plotting heap map between all the variables
sns.set(rc={'figure.figsize':(20,10)})
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

**Possible correlation patterns among variables. Especially observe the correlation between input variables and target variable. Here, chest pain(cp), thalach and slope are positively correlated to target when compared to other variables.**

```python
#Bar plot to notice the frequency of heart disease in age groups
dfta=data[data['target']==1]['age'].value_counts().sort_index()
ax=dfta.plot.bar(figsize=(15,6),fontsize=15)
ax.set_xlabel("Age", fontsize=15)
ax.set_ylabel("Frequency", fontsize=15)
ax.set_title("Frequency vs Age of heart disease", fontsize=15)
plt.show()
```

**Having a frequency threshold of 6, people between the age 40 to 60 are at high risk of heart disease.**

```python
#Bar plot to notice the frequency of heart disease between male and female
dfts=data[data['target']==1]['sex'].value_counts()
ax=dfts.plot.bar(figsize=(10,5),title='Frequency of heart disease in male(1) vs female(0)',fontsize=15)
ax.set_xlabel("sex", fontsize=15)
ax.set_ylabel("Frequency", fontsize=15)
ax.set_title('Frequency of heart disease in male(1) vs female(0)',fontsize=15)
plt.show()
```

**Here by concludes male are more affected by heart disease than female.**
